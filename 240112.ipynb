```python
df2=df1.head(10)
plt.bar(df2.index,df2['sepal_length'])
plt.rc('font', family='NanumGothic')
plt.xlabel('Data')
plt.ylabel('꽃받침의 길이')
plt.title('꽃받침 길이 시각화')
plt.show()
```

```python
plt.bar(['setosa','versicolor','virginica'],df1.groupby('species')['sepal_length'].mean())
plt.xlabel('종류')
plt.ylabel('꽃받침 길이의 평균값')
plt.title('품종별 sepal length 그래프')
plt.show()
```

```python
plt.hist(df1['sepal_length'],bins=3,label='Sepal Length')
plt.hist(df1['sepal_width'],bins=3,label='Sepal Width')
plt.hist(df1['petal_length'],bins=3,label='Petal Length')
plt.hist(df1['petal_width'],bins=3,label='Petal Width')
plt.legend()
plt.xlabel('Value')
plt.ylabel('빈도수')
plt.title('각 속성 값들의 빈도수 히스토그램')
```

```python
sns.scatterplot(x=df1['sepal_length'],y=df1['petal_length'])
```

```python
dff.describe()
```

```python
dff.sort_values(by=['fare'],ascending=False)
```

```python
dff['survived'].value_counts()
```

```python
dff.isna().sum()
```

```python
dff1=dff.drop('deck',axis=1)
dff1
```

```python
dff1.isna().sum()
```

```python
dff1.columns
```

```python
dff1.value_counts('embarked')
```

```python
dff2=dff1.fillna({'embarked':'S'})
dff2.isna().sum()
```

```python
age_avg=dff2['age'].mean()
dff3=dff2.fillna({'age':age_avg})
dff3.isna().sum()
```

```python
dff3
```

```python
dff3.groupby('embarked')['pclass'].mean()
```

```python
dff3.groupby(['pclass','sex']).mean()
```

```python
sns.countplot(x='pclass',data=dff3,hue='survived')
```

```python
a=dff3.corr()
sns.heatmap(data=a,annot=True,cbar=True,cmap='Blues')
```
